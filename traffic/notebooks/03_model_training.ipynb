```python
import sys
import os
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Add the parent directory to sys.path so we can import from 'src'
sys.path.append(os.path.abspath('../'))
%load_ext autoreload
%autoreload 2

from src.models.traffic_model_pipeline import TrafficModelPipeline

print("Setup completed successfully!")
print(f"Working directory: {os.getcwd()}")
```

## Initialize Enhanced Model Pipeline

Initialize the enhanced modeling pipeline with comprehensive ML capabilities.

```python
# Initialize the enhanced model pipeline
model_pipeline = TrafficModelPipeline(
    data_path="../data/processed/traffic_model_ready.csv",
    target="most_severe_injury"
)

print("Enhanced Model Pipeline initialized!")
print(f"Target variable: {model_pipeline.target}")
```

## Load Processed Data with Validation

Load the processed dataset with comprehensive validation and quality checks.

```python
# Load processed data with enhanced validation
df = model_pipeline.load_data()

if df is not None:
    print(f"\nDataset loaded successfully!")
    print(f"   Ready for model training with {df.shape[1]-1} features")
else:
    print("Failed to load data. Please check the data path.")
```

## Load Pre-processed Train/Test/Validation Splits

Load the train/test/validation splits created during data preparation phase.

```python
# Load pre-saved splits with validation set
try:
    splits = model_pipeline.load_splits_from_dir("../data/processed", include_validation=True)
    
    if len(splits) == 6:  # With validation set
        X_train, X_val, X_test, y_train, y_val, y_test = splits
        print(f"\nLoaded splits with validation set:")
        print(f"   Training: {X_train.shape[0]:,} samples")
        print(f"   Validation: {X_val.shape[0]:,} samples")
        print(f"   Testing: {X_test.shape[0]:,} samples")
    else:  # Without validation set
        X_train, X_test, y_train, y_test = splits
        X_val, y_val = None, None
        print(f"\nLoaded splits without validation set:")
        print(f"   Training: {X_train.shape[0]:,} samples")
        print(f"   Testing: {X_test.shape[0]:,} samples")
    
    print(f"   Features: {X_train.shape[1]}")
    
except Exception as e:
    print(f"Error loading splits: {e}")
    print("   Please ensure data preparation has been completed.")
```

## Define Core Model Suite

Define a focused suite of 3 essential machine learning algorithms:
- **Logistic Regression**: Linear probabilistic classifier
- **Decision Tree**: Rule-based interpretable model  
- **Random Forest**: Ensemble of decision trees

```python
# Define core model suite
models = model_pipeline.define_models()

print(f"\nCore Model Suite Defined:")
print(f"   Total models: {len(models)}")
print(f"   Focus on essential, interpretable algorithms")
print(f"   All models configured with balanced class weights")
```

## Cross-Validation Analysis

Perform comprehensive cross-validation to get robust performance estimates before final training.

```python
# Perform cross-validation analysis
cv_results = model_pipeline.perform_cross_validation(X_train, y_train, cv_folds=5)

print(f"\nCross-Validation Results Summary:")
print(cv_results[['Model', 'CV_Accuracy_Mean', 'CV_F1_Mean']].round(4))

# Identify top performers
best_cv_model = cv_results.loc[cv_results['CV_F1_Mean'].idxmax(), 'Model']
best_cv_score = cv_results['CV_F1_Mean'].max()
print(f"\nBest CV Performance: {best_cv_model} (F1: {best_cv_score:.4f})")
```

## Enhanced Model Training & Evaluation

Train all 3 core models and perform comprehensive evaluation with:
- **Advanced Metrics**: Accuracy, Precision, Recall, F1, Matthews Correlation, Cohen's Kappa
- **Confusion Matrices**: Detailed classification analysis
- **ROC Curves**: Multi-class ROC analysis
- **Feature Importance**: Model interpretability analysis
- **Training Time Tracking**: Performance monitoring

```python
# Train and evaluate all models with comprehensive analysis
results_df = model_pipeline.train_and_evaluate(
    X_train, X_test, y_train, y_test, X_val, y_val
)

print(f"\nTraining Complete!")
print(f"   Models trained: {len(results_df)}")
print(f"   Comprehensive visualizations generated")
print(f"   Feature importance analysis completed")

# Display results summary
print(f"\nModel Performance Summary:")
display_cols = ['Model', 'Accuracy', 'F1_Score', 'Precision', 'Recall', 'Training_Time']
available_cols = [col for col in display_cols if col in results_df.columns]
print(results_df[available_cols].round(4))
```

## Best Model Selection & Saving

Select and save the best performing model with comprehensive metadata.

```python
# Combine original and tuned results for final selection
all_results = results_df.copy()
# Select best model based on F1-Score
best_model_name = model_pipeline.get_best_model_name(all_results, metric="F1_Score")
best_model_score = all_results.loc[all_results['Model'] == best_model_name, 'F1_Score'].iloc[0]

print(f"Best Model Selected: {best_model_name}")
print(f"F1-Score: {best_model_score:.4f}")

# Save best model with metadata
model_path = model_pipeline.save_best_model(
    all_results,
    model_dir="../models",
    metric="F1_Score",
    include_tuned=True
)

print(f"\nModel saved successfully!")
print(f"   Model file: {model_path}")
print(f"   Metadata included for reproducibility")
```
